```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset
df = pd.read_parquet(r"C:\Users\rl587\PycharmProjects\Depth_to_groundwater\optimized_gwsi.parquet")

# Drop rows with missing values for comparison
df['WL_ELEV_CLEAN'] = df['WELL_ALT'] - df['WL_DTW']
df['DEPTH_BLW_LANDSURF'] =df['WL_ELEV_CLEAN'] - df['WELL_ALT']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Basic boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='AQ_NAME', y='WL_ELEV_CLEAN')
plt.xticks(rotation=90)
plt.title("Water Level Elevation by Geologic Unit (AQ_NAME)")
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans

for aq in df['AQ_NAME'].unique():
	
	aquifer = aq  # Replace with the desired aquifer name
	subset = df[df['AQ_NAME'] == aquifer].dropna(subset=['DEPTH_BLW_LANDSURF'])

	# Check if the subset is not empty
	if not subset.empty:
		# Try k-means with k=2 or 3 to start
		X = subset[['DEPTH_BLW_LANDSURF']].values
		kmeans = KMeans(n_clusters=5, random_state=0).fit(X)
		subset['cluster'] = kmeans.labels_

		# Visualize
		sns.histplot(data=subset, x='DEPTH_BLW_LANDSURF', hue='cluster', bins=30, kde=True)
		plt.title(f"Water Level Clustering in {aquifer}")
		plt.show()
	else:
		print(f"No data available for the aquifer: {aquifer}")
```

```python
import plotly.express as px

plot_df = df[['DD_LONG', 'DD_LAT', 'DEPTH_BLW_LANDSURF', 'AQ_NAME']].dropna()
plot_df = plot_df[plot_df['DEPTH_BLW_LANDSURF'] <= 2000]  # optional filter

fig = px.scatter_3d(
    plot_df,
    x='DD_LONG',
    y='DD_LAT',
    z='DEPTH_BLW_LANDSURF',
    color='AQ_NAME',
    title='3D Plot of Water Table Depths Below Land Surface',
    height=700,
    labels={
        'DD_LONG': 'Longitude',
        'DD_LAT': 'Latitude',
        'DEPTH_BLW_LANDSURF': 'Depth Below Land Surface (ft)'
    }
)

#  Flip Z-axis and set range
fig.update_layout(
    scene=dict(
        zaxis=dict(
            title='Depth Below Surface (ft)',
            autorange='reversed',
            range=[0, 2000]
        )
    )
)

#  Make points smaller
fig.update_traces(marker=dict(size=2))

fig.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count valid depth values per unit
valid_counts = df[['UNIT_NAME', 'DEPTH_BLW_LANDSURF']].dropna().groupby('UNIT_NAME').size()

# Filter to units with at least 10 valid entries
unit_names_with_enough_data = valid_counts[valid_counts >= 10].sort_values(ascending=False)

# Take top 15 most common units
top_units = unit_names_with_enough_data.head(15).index

# Subset the data
box_df = df[df['UNIT_NAME'].isin(top_units)][['UNIT_NAME', 'DEPTH_BLW_LANDSURF']].dropna()

# Plot
plt.figure(figsize=(14, 6))
sns.boxplot(data=box_df, x='UNIT_NAME', y='DEPTH_BLW_LANDSURF')
plt.xticks(rotation=90)
plt.title("Depth Below Land Surface by Top 15 Geologic Units (UNIT_NAME, 10 points)")
plt.ylabel("Depth Below Land Surface (ft)")
plt.xlabel("Geologic Unit")
plt.tight_layout()
plt.show()
```

```python
# Get medians
median_depths = (
    df[['UNIT_NAME', 'DEPTH_BLW_LANDSURF']]
    .dropna()
    .groupby('UNIT_NAME')['DEPTH_BLW_LANDSURF']
    .median()
)

# Keep only units with at least 10 points
valid_medians = median_depths[unit_names_with_enough_data.index]

# Sort by median depth
top_units_by_median = valid_medians.sort_values().head(15).index

# Filter for plotting
box_df = df[df['UNIT_NAME'].isin(top_units_by_median)][['UNIT_NAME', 'DEPTH_BLW_LANDSURF']].dropna()

plt.figure(figsize=(14, 6))
sns.boxplot(data=box_df, x='UNIT_NAME', y='DEPTH_BLW_LANDSURF', order=top_units_by_median)
plt.xticks(rotation=90)
plt.title("Shallowest 15 Geologic Units by Median Depth to Water (10 points)")
plt.ylabel("Depth Below Land Surface (ft)")
plt.xlabel("Geologic Unit")
plt.tight_layout()
plt.show()
```

```python
# Create label with count
label_map = {unit: f"{unit}\n(n={valid_counts[unit]})" for unit in top_units}
box_df['UNIT_NAME_LABELED'] = box_df['UNIT_NAME'].map(label_map)

plt.figure(figsize=(14, 6))
sns.boxplot(data=box_df, x='UNIT_NAME_LABELED', y='DEPTH_BLW_LANDSURF')
plt.xticks(rotation=90)
plt.title("Depth Below Land Surface by Top 15 Geologic Units (with sample sizes)")
plt.ylabel("Depth Below Land Surface (ft)")
plt.xlabel("Geologic Unit")
plt.tight_layout()
plt.show()
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python

# Get list of unique unit names, sorted alphabetically
unit_options = sorted(df['UNIT_NAME'].dropna().unique())

# Create dropdown widget
unit_dropdown = widgets.Combobox(
    options=unit_options,
    placeholder='Select a geologic unit...',
    description='Unit:',
    ensure_option=True,
    layout=widgets.Layout(width='50%')
)

# Function to update plot when dropdown changes
def update_boxplot_with_hist(change):
    clear_output(wait=True)
    display(unit_dropdown)
    
    selected_unit = change['new']
    subset = df[df['UNIT_NAME'] == selected_unit][['DEPTH_BLW_LANDSURF']].dropna()

    n = len(subset)
    if n == 0:
        print(f"No valid data for {selected_unit}")
        return

    # Create figure with two subplots (shared y-axis)
    fig, (ax_box, ax_hist) = plt.subplots(
        nrows=1, ncols=2, figsize=(8, 5), gridspec_kw={'width_ratios': [2, 1]}, sharey=True
    )

    # Boxplot
    sns.boxplot(data=subset, y='DEPTH_BLW_LANDSURF', ax=ax_box, color='red')
    ax_box.set_title(f"{selected_unit}\nDepth Below Land Surface\nn = {n}")
    ax_box.set_xlabel("")
    ax_box.set_ylabel("Depth (ft)")

    # Histogram (horizontal)
    ax_hist.hist(subset['DEPTH_BLW_LANDSURF'], bins=20, orientation='horizontal', color='red', alpha=0.7)
    ax_hist.set_xlabel("Count")
    ax_hist.set_ylabel("")  # Already on left axis
    ax_hist.grid(False)

    plt.tight_layout()
    plt.show()

# Attach handler to dropdown
unit_dropdown.observe(update_boxplot_with_hist, names='value')

# Display the widget
display(unit_dropdown)

```

```python


import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output

# Fields you can group by
group_fields = [
    'UNIT_NAME', 'AGE_MIN', 'AGE_MAX', 'MAJOR1', 'MAJOR2', 'MAJOR3',
    'MINOR1', 'MINOR2', 'MINOR3', 'GENERALIZE', 'ROCK_NAME',
    'ROCK_TYPE', 'AQ_NAME'
]

# Dropdown widgets
group_by_dropdown = widgets.Dropdown(
    options=group_fields,
    description='Group by:',
    layout=widgets.Layout(width='30%')
)

value_dropdown = widgets.Dropdown(
    options=[],
    description='Value:',
    layout=widgets.Layout(width='40%')
)

# Layout container
ui = widgets.HBox([group_by_dropdown, value_dropdown])

# Update value dropdown options based on selected group field
def update_value_options(change):
    field = change['new']
    unique_vals = sorted(df[field].dropna().unique())
    value_dropdown.options = unique_vals

# Update plots when value is selected
def update_plot(change):
    clear_output(wait=True)
    display(ui)

    field = group_by_dropdown.value
    value = value_dropdown.value

    if not field or not value:
        return

    subset = df[
        (df[field] == value) &
        (df['DEPTH_BLW_LANDSURF'].notna())
    ]

    n = len(subset)
    if n == 0:
        print(f"No valid data for {field} = {value}")
        return

    stats = subset['DEPTH_BLW_LANDSURF'].describe()
    iqr = stats['75%'] - stats['25%']

    # Random color
    random_color = np.random.rand(3,)

    # Plot with 3 subplots: boxplot, histogram, and text panel
    fig, (ax_box, ax_hist, ax_text) = plt.subplots(
        nrows=1, ncols=3, figsize=(12, 5),
        gridspec_kw={'width_ratios': [2, 2, 1]}, sharey=True
    )

    # Boxplot
    sns.boxplot(data=subset, y='DEPTH_BLW_LANDSURF', ax=ax_box, color=random_color)
    ax_box.set_title(f"{field} = {value}\nn = {n}")
    ax_box.set_xlabel("")
    ax_box.set_ylabel("Depth Below Land Surface (ft)")

    # Histogram
    ax_hist.hist(subset['DEPTH_BLW_LANDSURF'], bins=20, orientation='horizontal', color=random_color, alpha=0.7)
    ax_hist.set_xlabel("Count")
    ax_hist.set_ylabel("")
    ax_hist.grid(False)

    # Text-based summary stats
    stat_lines = [
        f"n       = {n}",
        f"Mean    = {stats['mean']:.2f} ft",
        f"Median  = {stats['50%']:.2f} ft",
        f"Min     = {stats['min']:.2f} ft",
        f"Max     = {stats['max']:.2f} ft",
        f"IQR     = {iqr:.2f} ft",
        f"Std Dev = {stats['std']:.2f} ft"
    ]

    ax_text.axis('off')
    ax_text.text(0, 1, "\n".join(stat_lines), fontsize=12, verticalalignment='top', family='monospace')

    plt.tight_layout()
    plt.show()

# Hook up the dropdown changes
group_by_dropdown.observe(update_value_options, names='value')
value_dropdown.observe(update_plot, names='value')

# Trigger initial value list
update_value_options({'new': group_fields[0]})

# Display UI
display(ui)
```
